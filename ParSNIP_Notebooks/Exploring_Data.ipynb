```python
transient_dirs = []
transient_dict = {}

for dirname, dirnames, filenames in os.walk(path):
    for subdirname in dirnames:
        print(subdirname)
        if subdirname[-2:].isdigit():
            transient_dirs.append(os.path.join(dirname, subdirname))
            transient_dict[subdirname] = {}
```

```python
for dir in transient_dirs: 
    objID = dir.rsplit('/', 1)[-1]
    for file in glob.glob(dir+'/photometry_*.csv'):
        data = pd.read_csv(file)
        data.rename(columns = {'filter':'band'}, inplace = True)
        data.rename(columns = {'mjd':'time'}, inplace = True)
        
    try:
        transient_dict[objID]['data'] = data 
    except KeyError as msg:
        print(msg)

```
